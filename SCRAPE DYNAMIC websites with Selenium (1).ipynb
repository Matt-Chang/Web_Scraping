```python
from selenium import webdriver 
import pandas as pd
```

```python
url = 'https://www.youtube.com/@KenJee_ds/videos'
```

```python
driver = webdriver.Chrome()

driver.get(url)
```

```python



lit =[]

videos = driver.find_elements(By.CLASS_NAME, "style-scope ytd-rich-grid-media")

for video in videos:
    title = video.find_elements(By.XPATH, './/*[@id="video-title-link"]')
    views = video.find_elements(By.XPATH, './/*[@id="metadata-line"]/span[1]')
    when = video.find_elements(By.XPATH, './/*[@id="metadata-line"]/span[2]')
    print(title[0].text,views[0].text,when[0].text)
    data = {
        "title":title[0].text,
        "views":views[0].text,
        "when":when[0].text
    }
    lit.append(data)
    
    
df = pd.DataFrame(lit)

print(df)
```
